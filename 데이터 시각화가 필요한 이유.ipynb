# 데이터 시각화
## 데이터 시각화가 필요한 이유
### 필요 라이브러리 불러오기

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### anscombe 데이터 불러오기

```python
anscombe = sns.load_dataset("anscombe")
anscombe
```

### anscombe 데이터의 dataset별 평균과 표준편차 알아보기

```python
anscombe.groupby("dataset").agg(['mean', 'std'])
```

### 조건별 데이터 분할하기

```python
data1 = anscombe[anscombe['dataset'] == 'I']
data2 = anscombe[anscombe['dataset'] == 'II']
data3 = anscombe[anscombe['dataset'] == 'III']
data4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
data1
```

### 데이터 시각화 준비하기

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
```

### 그림이 그려질 자리 할당하기

```python
ax1.scatter(data1['x'], data1['y'])
ax2.scatter(data2['x'], data2['y'])
ax3.scatter(data3['x'], data3['y'])
ax4.scatter(data4['x'], data4['y'])

fig
```

### 그림별 제목 설정하기

```python
ax1.set_title("Data1")
ax2.set_title("Data2")
ax3.set_title("Data3")
ax4.set_title("Data4")

fig
```

### 메인 제목 설정하기

```python
fig.suptitle("Anscombe Data")
fig
```

### 그림 레이아웃 맞추기

```python
fig.tight_layout()
fig
```
